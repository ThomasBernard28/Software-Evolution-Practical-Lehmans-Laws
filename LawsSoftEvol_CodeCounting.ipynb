### Dataset A : Major.Minor.Patch tag format

Filter out non-official releases such as alpha, beta, release candidates, milestones, branches, and unnumbered tags. Additionally, version numbers typically follow a Major.Minor.Patch format, but exceptions exist (e.g., 1.4, 3.3.0.1, 4.3.0.1, or 4.3.1.1). Versions with more than three components are disregarded, and two-component versions are treated as implicit three-component versions (e.g., 1.4 is interpreted as 1.4.0).  

Write a regular expression statement to identify tags following semantic versioning (major.minor.patch) guidelines. You can use the pattern: '^v?[0-9]+\.[0-9]+(\.[0-9]+)?$'  

Once you've applied this regex, print out the selected major.minor.patch dataset.

```python
# WRITE YOUR CODE HERE
filtered_data = growth_data[growth_data['tag'].str.contains('^v?[0-9]+.[0-9]+(.[0-9]+)?$')]

print(filtered_data)
```

### Dataset B : Major.Minor tag format

Starting from the previous data (**Dataset A**),  for each Major.Minor.Patch tag group, select the earliest version (e.g., for a set of 4.4.0, 4.4.1, and 4.4.2, consider 4.4.0 and rename it to 4.4).

```python
# WRITE YOUR CODE HERE
versioned_df = filtered_data.copy(deep=True)
versioned_df['Major_Minor'] = versioned_df['tag'].apply(lambda x: '.'.join(x.split('.')[:2]))
earliest_versions = versioned_df.groupby('Major_Minor', as_index=False).first()
earliest_versions.drop(columns=['Major_Minor'], inplace=True)
earliest_versions['tag'] = earliest_versions['tag'].str.extract(r'^(\d+.\d+)')
print(earliest_versions)
    
```

### Regression analysis

Visualize a linear regression analysis of the relationship between the release date and the total number of files by considering **Dataset B**.  

Hint:

1. Prepare the data: Convert the release date to integers and set number of files as the target variable.  
2. Apply linear regression analysis to understand the relationship between the release date and the total number of files.  
3. Calculate Mean Relative Error and ajusted R-squared metrics.    
4. Generate a scatter plot showing the release dates against the total number of files. Then, overlay the linear regression line on the plot.


```python
# WRITE YOUR CODE HERE
filtered_data['release_date'] = pd.to_datetime(filtered_data['release_date']).astype(int)
filtered_data
y = filtered_data['nFiles']
x = filtered_data[['release_date']]
lr = LinearRegression()
lr.fit(x, y)
y_pred = lr.predict(x)
mre = np.mean(np.abs((y - y_pred) / y)) * 100

n = len(y)
p = 1 
r_squared = lr.score(x, y)
adjusted_r2 = 1 - (1 - r_squared) * (n - 1) / (n - p - 1)

print(f"Mean Relative Error: {mre:.3f}")
print(f"Adjusted R2: {adjusted_r2:.3f}")

filtered_data['release_date'] = pd.to_datetime(filtered_data['release_date'])

plt.scatter(filtered_data['release_date'], filtered_data['nFiles'], label='Actual Data')
plt.plot(filtered_data['release_date'], y_pred, color='red', label='Linear Regression')
plt.xlabel('Release Date')
plt.ylabel('Number of Files')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# WRITE YOUR CODE HERE
filtered_data['release_date'] = pd.to_datetime(filtered_data['release_date'])
# earliest release date
earliest_date = filtered_data['release_date'].min()
filtered_data['release_days'] = (filtered_data['release_date'] - earliest_date).dt.days
filtered_data
```

```python
# WRITE YOUR CODE HERE
y = filtered_data['nFiles']


split_date = '2015-07-28'
train_data = filtered_data[filtered_data['release_date'] < split_date]
test_data = filtered_data[filtered_data['release_date'] >= split_date]
```

```python
# WRITE YOUR CODE HERE
# Define regression models
lr = LinearRegression()
lr.curve_fit = True
#fit model to training data 
lr.fit(train_data[['release_days']], train_data['nFiles'])

```

```python
import sklearn.preprocessing
x = train_data[['release_days']]

poly_features = sklearn.preprocessing.PolynomialFeatures(degree=2)
x_poly = poly_features.fit_transform(x)

quad_reg = LinearRegression()
quad_reg.fit(x_poly, train_data['nFiles'])
```

```python
# make a logarithmic regression model
y_train = np.log(train_data['nFiles'])
y_test = np.log(test_data['nFiles'])
log_reg = LinearRegression()
log_reg.fit(x, y_train)
intercept = log_reg.intercept_
slope = log_reg.coef_[0]
```

```python
# WRITE YOUR CODE HERE
y_pred = lr.predict(test_data[['release_days']])

y_quad_pred = quad_reg.predict(poly_features.fit_transform(test_data[['release_days']]))

y_log_train_pred = np.exp(intercept + slope * x)
y_log_test_pred = np.exp(intercept + slope * test_data[['release_days']])
```

```python
# WRITE YOUR CODE HERE
#plot the actual data and the predicted data
plt.scatter(filtered_data['release_days'], filtered_data['nFiles'], label='Actual Data')
plt.plot(test_data['release_days'], y_pred, color='red', label='Linear Regression')
plt.scatter(test_data['release_days'], y_pred, color='red', label='Predicted Data')
plt.xlabel('Release Days')
plt.ylabel('Number of Files')
plt.title('Linear Regression')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
